```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly as py
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
athlete_bio = pd.read_csv('/content/drive/MyDrive/Datasets/Olympic_Athlete_Bio.csv')
medal_tally = pd.read_csv('/content/drive/MyDrive/Datasets/Olympic_Games_Medal_Tally.csv')
results = pd.read_csv('/content/drive/MyDrive/Datasets/Olympic_Results.csv')
athlete_event_results = pd.read_csv('/content/drive/MyDrive/Datasets/Olympic_Athlete_Event_Results.csv')
games = pd.read_csv('/content/drive/MyDrive/Datasets/Olympics_Games.csv')
countries = pd.read_csv('/content/drive/MyDrive/Datasets/Olympics_Country.csv')
```

```python
merged_df = athlete_event_results.merge(athlete_bio, on='athlete_id', how='outer', suffixes=('', '_DROP')).filter(regex='^(?!.*_DROP)')
merged_df = merged_df.merge(games, on='edition_id', how='outer', suffixes=('', '_DROP')).filter(regex='^(?!.*_DROP)')
summer_df = merged_df[merged_df['edition'].fillna('').str.contains('Summer')]
winter_df = merged_df[merged_df['edition'].fillna('').str.contains('Winter')]
```

# **Number of Countries in Summer Olympics**

```python
country_count = (summer_df.groupby(['year', 'country_noc']).size()).groupby('year').size().reset_index(name='country_count')
country_count.head()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=country_count, x='year', y='country_count')
plt.title('Number of Countries Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Countries')
plt.show()
```

# **Top 10 Countries with Most Number of Medals in Summer Olympics**

```python
summer_medal_count = medal_tally[medal_tally['edition'].str.contains('Summer', na=False)]
summer_medal_count = (summer_medal_count
                       .groupby(['country_noc', 'country'])
                       .agg({'total': 'sum', 'gold': 'sum', 'silver': 'sum', 'bronze': 'sum'})
                       .reset_index()
                       .sort_values(by='total', ascending=False)
                       .head(10))
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='total', y='country', data=summer_medal_count, palette='viridis')

plt.title('Total Medals Won by Country')
plt.xlabel('Total Medals')
plt.ylabel('Country')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.barh(summer_medal_count['country'], summer_medal_count['bronze'], color='brown', label='Bronze')
plt.barh(summer_medal_count['country'], summer_medal_count['silver'], left=summer_medal_count['bronze'], color='gray', label='Silver')
plt.barh(summer_medal_count['country'], summer_medal_count['gold'], left=summer_medal_count['bronze'] + summer_medal_count['silver'], color='gold', label='Gold')

plt.title('Top 10 Countries by Medal Type')
plt.xlabel('Total Medals')
plt.ylabel('Country')
plt.legend()

plt.show()
```

# **Top 10 Countries with Most Number of Medals in Winter Olympics**

```python
winter_medal_count = medal_tally[medal_tally['edition'].str.contains('Winter', na=False)]
winter_medal_count = (winter_medal_count.groupby(['country_noc', 'country'])
                       .agg({'total': 'sum', 'gold': 'sum', 'silver': 'sum', 'bronze': 'sum'})
                       .reset_index()
                       .sort_values(by='total', ascending=False)
                       .head(10))
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='total', y='country', data=winter_medal_count, palette='viridis')

plt.title('Total Medals Won by Country')
plt.xlabel('Total Medals')
plt.ylabel('Country')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.barh(winter_medal_count['country'], winter_medal_count['bronze'], color='brown', label='Bronze')
plt.barh(winter_medal_count['country'], winter_medal_count['silver'], left=winter_medal_count['bronze'], color='gray', label='Silver')
plt.barh(winter_medal_count['country'], winter_medal_count['gold'], left=winter_medal_count['bronze'] + winter_medal_count['silver'], color='gold', label='Gold')

plt.title('Top 10 Countries by Medal Type')
plt.xlabel('Total Medals')
plt.ylabel('Country')
plt.legend()

plt.show()
```

# **Top 10 Athletes with Most Number of Medals in Summer Olympics**
